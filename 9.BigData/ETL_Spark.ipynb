# Persistimos nuestros DF en una tabla


- Primero vamos a avisarle a usar la API de SQL de spark para hacer las cosas mas simples
- Luego vamos a guardar nuestro DF en HDFS y vamos a crear una tabla

```
#crear un nuevo daframe joinedDF joineando los dataframes gameInfDf y eventsDf con los siguientes campos:
# eventsDf.id_odsp, eventsDf.id_event, eventsDf.sort_order, eventsDf.time, eventsDf.event_type, eventsDf.event_type_str, eventsDf.event_type2, eventsDf.event_type2_str, eventsDf.side, eventsDf.side_str, eventsDf.event_team, eventsDf.opponent, eventsDf.player, eventsDf.player2, eventsDf.player_in, eventsDf.player_out, eventsDf.shot_place, eventsDf.shot_place_str, eventsDf.shot_outcome, eventsDf.shot_outcome_str, eventsDf.is_goal, eventsDf.location, eventsDf.location_str, eventsDf.bodypart, eventsDf.bodypart_str, eventsDf.assist_method, eventsDf.assist_method_str, eventsDf.situation, eventsDf.situation_str, gameInfDf.country_code 

sqlContext.registerDataFrameAsTable(eventsDf, "eventsDf")  # para que spark pueda usar la API de SQL, vamos a tener que registrar nuestro Dataframe como tabla
sqlContext.registerDataFrameAsTable(gameInfDf, "gameInfDf")

joinedDf = sqlContext.sql( # usamos sparkSQL
  """select eventsDf.id_odsp, eventsDf.id_event, eventsDf.sort_order, eventsDf.time, eventsDf.event_type, eventsDf.event_type_str,
            eventsDf.event_type2, eventsDf.event_type2_str, eventsDf.side, eventsDf.side_str, eventsDf.event_team, eventsDf.opponent, 
            eventsDf.player, eventsDf.player2, eventsDf.player_in, eventsDf.player_out, eventsDf.shot_place, eventsDf.shot_place_str,
            eventsDf.shot_outcome,eventsDf.shot_outcome_str, eventsDf.is_goal, eventsDf.location, eventsDf.location_str, eventsDf.bodypart,
            eventsDf.bodypart_str,eventsDf.assist_method, eventsDf.assist_method_str, eventsDf.situation, eventsDf.situation_str,
            gameInfDf.league, gameInfDf.country_code    
     FROM eventsDf
     INNER JOIN gameInfDf 
       ON eventsDf.id_odsp == gameInfDf.id_odsp
 """)
joinedDf.show()
```
