## Preparación de datos 



```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import operator
```

Creamos una primera función llamada cambiar_a_flotante que recibe un número y lo convierte a float(). Si el parámetro está vacío, lo iguala a 0. 

```python
def cambiar_a_flotante (num): 
  try: 
    num = float(num)
  except:
    num = 0
  return num
```

Creamos la función lee_archivo_csv_expectativa_de_vida. Utilizamos la misma para que, a partir de un archivo CSV, llene un diccionario con los distritos como claves y la información pertinente de cada uno como valor en un nuevo diccionario, donde la clave será la columna y el valor sus valores. 

```python
def lee_archivo_csv_expectativa_de_vida (archivo_csv, diccionario): 

  try: 
    with open (archivo_csv, mode="r") as csv_file: 
      csv_reader = csv.DictReader (csv_file, delimiter = ";")
      for row in csv_reader:  
        if str(row["District Name"]) not in diccionario:
          diccionario[str(row["District Name"])] = {           
            "Neighborhood": [row["Neighborhood"]], 
            "2006-2010": [cambiar_a_flotante(row["2006-2010"])],
            "2007-2011": [cambiar_a_flotante(row["2007-2011"])],
            "2008-2012": [cambiar_a_flotante(row["2008-2012"])],
            "2009-2013": [cambiar_a_flotante(row["2009-2013"])],
            "2010-2014": [cambiar_a_flotante(row["2010-2014"])],
            "Gender": [row["Gender"]],  
        }

        else: 
          diccionario.get(str(row["District Name"])).get("Neighborhood").append(cambiar_a_flotante(row["Neighborhood"]))
          diccionario.get(str(row["District Name"])).get("2006-2010").append(cambiar_a_flotante(row["2006-2010"])) 
          diccionario.get(str(row["District Name"])).get("2007-2011").append(cambiar_a_flotante(row["2007-2011"])) 
          diccionario.get(str(row["District Name"])).get("2008-2012").append(cambiar_a_flotante(row["2008-2012"]))
          diccionario.get(str(row["District Name"])).get("2009-2013").append(cambiar_a_flotante(row["2009-2013"]))
          diccionario.get(str(row["District Name"])).get("2010-2014").append(cambiar_a_flotante(row["2010-2014"]))           
          diccionario.get(str(row["District Name"])).get("Gender").append(row["Gender"])
        
  except FileNotFoundError: 
    print("No se inicializó el archivo")
    
diccionario_expectativa_de_vida = {}
lee_archivo_csv_expectativa_de_vida(csv_expectativa_de_vida, diccionario_expectativa_de_vida)
diccionario_expectativa_de_vida
```

Creamos la función lee_archivo_csv_nacimientos. Utilizamos la misma para que, a partir de un archivo CSV, llene un diccionario con los distritos como claves y la información pertinente de cada uno como valor en listas. 

```python

def lee_archivo_csv_nacimientos(archivo_csv, diccionario):   
  try: 
    with open(archivo_csv, mode="r") as csv_file: 
      csv_reader = csv.DictReader(csv_file, delimiter = ";")
      for row in csv_reader:  
        if (str(row["District Name"])) not in diccionario:
          diccionario[str(row[ "District Name"])] = {           
            "Year": [cambiar_a_flotante(row["Year"])],
            "Neighborhood Code": [cambiar_a_flotante(row["Neighborhood Code"])],
            "Neighborhood Name": [row["Neighborhood Name"]],
            "Gender": [row["Gender"]],
            "Number": [cambiar_a_flotante(row["Number"])],
            }
        else:
         diccionario.get(str(row["District Name"])).get("Year").append(cambiar_a_flotante(row["Year"]))
         diccionario.get(str(row["District Name"])).get("Neighborhood Code").append(row["Neighborhood Code"])
         diccionario.get(str(row["District Name"])).get("Neighborhood Name").append(row["Neighborhood Name"])
         diccionario.get(str(row["District Name"])).get("Gender").append(row["Gender"])
         diccionario.get(str(row["District Name"])).get("Number").append(cambiar_a_flotante(row["Number"]))

  except FileNotFoundError: 
    print("No se inicializó el archivo")

diccionario_nacimientos = {}
lee_archivo_csv_nacimientos(csv_nacimientos, diccionario_nacimientos)
diccionario_nacimientos
```

## Análisis



Creamos una función llamada calcula_promedio_natalidad que itera por cada uno los distritos y crea un arreglo donde se lo corresponde con su promedio de natalidad. 

```python
def calcula_promedio_natalidad(diccionario_nacimientos):
  lista = [] 
  dic = {}
  
  for key in diccionario_nacimientos.keys():
    lista.extend(diccionario_nacimientos[key]["Number"])
    array = np.array(lista)
    promedio = np.average(array)
    dic[key] = float(" %.4f" %promedio)

  return dic

diccionario_promedio_nacimientos = calcula_promedio_natalidad(diccionario_nacimientos)
diccionario_promedio_nacimientos
```

## Visualizaciones



```python
#Expectativa de vida por genero

etiquetas = diccionario_promedios_generos.keys()
valores_dic = list(diccionario_promedios_generos.values())
hombres=[]
mujeres=[]
hombres = np.array(hombres, dtype = np.float32)
mujeres = np.array(mujeres, dtype = np.float32)
for lista in valores_dic:
  hombres = np.append(hombres,lista[0])
  mujeres = np.append(mujeres,lista[1])


list(hombres)
list(mujeres)

x = np.arange(len(etiquetas))  # La ubicacion de las etiquetas
width = 0.35  # El ancho de las barras

fig, ax = plt.subplots()
 
ax.bar(x - width/2, hombres, width, label='Hombres')
ax.bar(x + width/2, mujeres, width, label='Mujeres',color='pink')
# Agregado de texto, etiquetas del grafico y nombre del mismo
ax.set_ylabel('Años')
ax.set_title('Expectativa de vida por Distrito por genero')
ax.set_xticks(x)
ax.set_xticklabels(etiquetas)
ax.legend()


fig.tight_layout()
plt.gcf().set_size_inches(15, 5)
plt.show()
```
